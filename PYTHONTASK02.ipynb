```python
#3
Tuple = (10,5,8,4,6,3,4,2,2,1)
def countx(Tuple, x):
    count = 0
    for ele in Tuple:
        if (ele == x):
            count = count + 1
    return count


x = 2
y = 4
z = 6
print(countx(Tuple, x))
print(countx(Tuple, y))
print(countx(Tuple, z))
```

```python
#4
Tuple = [('arifa taj',21) , ('fida',22) , ('sabitha',20) , ('misna',23)]
def convert(tup, di):
    for a, b in tup:
        di.setdefault(a, []).append(b)
    return di
dic = {}
print(convert(Tuple, dic))
```
